```python
import random
import time
from insertion_sort import insertion_sort
from selection_sort import selection_sort
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)

average_case_insertion_sort = []
best_case_insertion_sort = []
worst_case_insertion_sort = []
average_case_selection_sort = []
best_case_selection_sort = []
worst_case_selection_sort = []

# random value
for n in range(1000, 15000, 1000):
    A = [random.randint(0, 1000) for _ in range(n)]
    start = time.time()
    insertion_sort(A.copy())
    end = time.time()
    average_case_insertion_sort.append([n, end - start])

    start = time.time()
    selection_sort(A.copy())
    end = time.time()
    average_case_selection_sort.append([n, end - start])


# best case and worst case test graph for insertion sort
for n in range(1000, 15000, 1000):
    A = [i for i in range(n)]
    start = time.time()
    insertion_sort(A.copy())
    end = time.time()
    best_case_insertion_sort.append([n, end - start])

    start = time.time()
    selection_sort(A.copy())
    end = time.time()
    best_case_selection_sort.append([n, end - start])

    A = [i for i in range(n, 0, -1)]
    start = time.time()
    insertion_sort(A)
    end = time.time()
    worst_case_insertion_sort.append([n, end - start])

    start = time.time()
    selection_sort(A)
    end = time.time()
    worst_case_selection_sort.append([n, end - start])

df1 = pd.DataFrame(average_case_insertion_sort, columns=['n', 'time'])
df2 = pd.DataFrame(best_case_insertion_sort, columns=['n', 'time'])
df3 = pd.DataFrame(worst_case_insertion_sort, columns=['n', 'time'])

df4 = pd.DataFrame(average_case_selection_sort, columns=['n', 'time'])
df5 = pd.DataFrame(best_case_selection_sort, columns=['n', 'time'])
df6 = pd.DataFrame(worst_case_selection_sort, columns=['n', 'time'])
```

```python
plt.title("Insertion Sort")
plt.xlabel("Number of data")
plt.ylabel("Time in seconds")
plt.plot(df1['n'], df1['time'], label="Average", linestyle='dotted')
plt.plot(df2['n'], df2['time'], label="Best", linestyle='solid')
plt.plot(df3['n'], df3['time'], label="Worst", linestyle='dashed')
plt.legend()
plt.show()
```

```python
plt.title("Selection Sort")
plt.xlabel("Number of data")
plt.ylabel("Time in seconds")
plt.plot(df4['n'], df4['time'], label="Average", linestyle='dotted')
plt.plot(df5['n'], df5['time'], label="Best", linestyle='solid')
plt.plot(df6['n'], df6['time'], label="Worst", linestyle='dashed')
plt.legend()
plt.show()
```
